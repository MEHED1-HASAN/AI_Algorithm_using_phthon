```python
import heapq

def A_star(graph, start, goal, heuristic):
    # Priority queue: (f = g+h, g, node, path)
    queue = [(heuristic[start], 0, start, [start])]
    visited = set()

    print("A* Search:")

    while queue:
        f, g, current, path = heapq.heappop(queue)
        print(f"{current}(g={g}, h={heuristic[current]}, f={f})", end=" ")

        if current == goal:
            print(f"\nGoal found with cost {g}")
             print("Path:", " → ".join(path))
            return

        if current not in visited:
            visited.add(current)

            for neighbor, cost in graph[current]:
                if neighbor not in visited:
                    g_new = g + cost
                    f_new = g_new + heuristic[neighbor]
                    new_path = path + [neighbor]
                    heapq.heappush(queue, (f_new, g_new, neighbor, new_path))

    print("\nGoal not found")


# Example
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}

heuristic = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 1,
    'E': 1,
    'F': 0   # Goal node always 0
}

start = 'A'
goal = 'F'

A_star(graph, start, goal, heuristic)
```
