```python
def UCS(graph, start, goal):
    queue = [(0, start)] 
    visited = set()
    
    print("Uniform Cost Search:")
    
    while queue:
        queue.sort()  
        cost, current = queue.pop(0)
        print(f"{current}({cost})", end=" ")
        
        if current == goal:
            print(f"\nGoal found with cost {cost}")
            return
            
        if current not in visited:
            visited.add(current)
            
            for neighbor, edge_cost in graph[current]:
                if neighbor not in visited:
                    new_cost = cost + edge_cost
                    queue.append((new_cost, neighbor))
    
    print(f"\nGoal not found")
    return 

print("Enter the graph with costs (example: {'A':[('B',1),('C',3)], 'B':[('D',2)], 'C':[('F',2)], 'D':[], 'F':[]}):")
graph = eval(input())
start = input("Enter start node: ")
goal = input("Enter goal node: ")

UCS(graph, start, goal)
```

```python
def UCS(graph, start, goal):
    queue = [(0, start, [start])]  # (cost, node, path)
    visited = set()
    
    print("Uniform Cost Search:")
    
    while queue:
        queue.sort()  # Sort by cost
        cost, current, path = queue.pop(0)
        print(f"{current}({cost})", end=" ")
        
        if current == goal:
            print(f"\nGoal found with cost {cost}")
            print(f"Path: {' → '.join(path)}")
            return
            
        if current not in visited:
            visited.add(current)
            
            for neighbor, edge_cost in graph[current]:
                if neighbor not in visited:
                    new_cost = cost + edge_cost
                    new_path = path + [neighbor]  # Extend the path
                    queue.append((new_cost, neighbor, new_path))
    
    print(f"\nGoal not found")

print("Enter the graph with costs:")
graph = eval(input())
start = input("Enter start node: ")
goal = input("Enter goal node: ")

UCS(graph, start, goal)
```
