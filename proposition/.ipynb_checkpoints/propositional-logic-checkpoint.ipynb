```python
def logic_evaluator():
    # Take two input statements
    stmt1 = input("Enter first statement (e.g., 'Raju is Doctor'): ")
    stmt2 = input("Enter second statement (e.g., 'Raju is Honest'): ")
    
    # Ask user for truth values in a simple way
    P = input(f"Is '{stmt1}' True or False? ") == "True"
    Q = input(f"Is '{stmt2}' True or False? ") == "True"

    
    # Ask user for operator
    operator = input("Enter logical operator (negation, conjunction, disjunction, implication, biconditional): ").lower()
    
    result = None
    output_sentence = ""
    
    if operator == "negation":
        result = not P
        output_sentence = f"~ ({stmt1})"
    
    elif operator == "conjunction":
        result = P and Q
        output_sentence = f"({stmt1}) ^ ({stmt2})"
    
    elif operator == "disjunction":
        result = P or Q
        output_sentence = f"({stmt1}) V ({stmt2})"
    
    elif operator == "implication":
        result = (not P) or Q
        output_sentence = f"({stmt1}) ---> ({stmt2})"
    
    elif operator == "biconditional":
        result = (P == Q)
        output_sentence = f"({stmt1}) <---> ({stmt2})"
    
    else:
        print("Invalid operator!")
        return
    
    print("\nResulting Expression: ", output_sentence)
    print("Truth Value: ", result)


# Run
logic_evaluator()
```

```python
def logic_evaluator():
    # Input statements
    stmt1 = input("Enter first statement: ")
    stmt2 = input("Enter second statement: ")
    
    # Input operator
    operator = input("Enter operator (negation, conjunction, disjunction, implication, biconditional): ").lower()
    
    values = [True, False]
    
    print("\nTruth Table:")
    print(f"{stmt1:25} {stmt2:25} Result")
    
    for P in values:
        for Q in values:
            if operator == "negation":
                result = not P
            elif operator == "conjunction":
                result = P and Q
            elif operator == "disjunction":
                result = P or Q
            elif operator == "implication":
                result = (not P) or Q
            elif operator == "biconditional":
                result = P == Q
            else:
                print("Invalid operator!")
                return
            
            print(f"{str(P):25} {str(Q):25} {str(result)}")

# Run
logic_evaluator()
```
