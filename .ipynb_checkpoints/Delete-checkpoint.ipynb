```python
def IDDFS(graph,start,goal,maxd):
    for i in range(maxd+1):
        visited=set()
        stack=[]
        visited.add(start)
        stack.append((start,0))
        print(f"\n DFS for Depth {i}:")
        

        while stack:
            current,depth=stack.pop()
            print(current, end=" ")
            if current==goal:
                print("\nGoal is found at level",depth)
                return 
            if depth<i:
                for neighbour in reversed(graph[current]):
                    if neighbour not in visited:
                        visited.add(neighbour)
                        stack.append((neighbour,depth+1))
        print("\nGoal is not found at level",i)
    print("\nGoal is not found in the Maximum limit")
    return

print("Enter the graph:")
graph=eval(input())
start=input("Enter the start node:")
goal=input("Enter the goal node:")
maxd=int(input("Enter the maximum depth:"))

IDDFS(graph,start,goal,maxd)
                
```
