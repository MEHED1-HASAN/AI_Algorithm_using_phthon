```python
from collections import deque

def simple_dfs(graph, start):
    visited = set()
    stack = deque()
    
    stack.append(start)
    
    print("DFS Traversal:")

    while stack:
        current_node = stack.pop()
        if current_node not in visited:
            print(current_node, end=" ")
            visited.add(current_node)

            # Add neighbors in reverse order for consistent left-to-right traversal
            for neighbor in reversed(graph[current_node]):
                if neighbor not in visited:
                    stack.append(neighbor)

print("Enter the graph in Python dictionary format (e.g., {'A': ['B', 'C'], 'B': ['D']}):")
graph = eval(input())
simple_dfs(graph, 'A')
```

```python
from collections import deque

def simple_dfs(graph, start):
    visited = set()
    stack = []   # use list as stack
    
    stack.append(start)
    visited.add(start)
    
    print("DFS Traversal:")

    while stack:
        current_node = stack.pop()   # LIFO
        print(current_node, end=" ")
        
        # add neighbours in reverse order for consistent left-to-right traversal
        for neighbor in reversed(graph[current_node]):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)

print("Enter the graph in Python dictionary format (e.g., {'A':['B','C'], 'B':['D']}):")
graph = eval(input())
simple_dfs(graph, 'A')
```
