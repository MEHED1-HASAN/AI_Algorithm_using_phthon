```python
from nltk import ngrams
from collections import Counter
import re

text = """
Machine learning is a branch of artificial intelligence that enables computers 
to learn from data and make predictions without being explicitly programmed. 
Deep learning uses neural networks with multiple layers for complex tasks.
"""

text=re.sub(r'[^a-zA-Z\s]','',text.lower())
tokens=text.split()

unigrams=list(ngrams(tokens,1))
bigrams=list(ngrams(tokens,2))
trigrams=list(ngrams(tokens,3))

print("Tokens:",tokens)
print("Unigrams:",unigrams[:4])
print("Bigrams:",bigrams[:4])
print("Trigrams:",trigrams[:3])

print("Top Unigrams:",Counter(unigrams).most_common(3))
print("Top Bigrams:",Counter(bigrams).most_common(3))
print("Top Trigrams:",Counter(trigrams).most_common(2))
```
