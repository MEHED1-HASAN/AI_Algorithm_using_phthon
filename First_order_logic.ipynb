```python
import itertools

# Dictionary to hold mapping between variables and sentences
sentences = {}

# Function to evaluate logical expression
def evaluate_expression(expr, values):
    expr_eval = expr.replace("^", " and ").replace("v", " or ").replace("~", " not ")
    for var, val in values.items():
        expr_eval = expr_eval.replace(var, str(val))
    return eval(expr_eval)

# Function to generate truth table
def truth_table(expr, variables):
    print(f"\nTruth Table for: {expr}")
    print(" | ".join(variables) + " | Result")
    print("-" * (len(variables) * 4 + 9))
    
    for combo in itertools.product([True, False], repeat=len(variables)):
        values = dict(zip(variables, combo))
        result = evaluate_expression(expr, values)
        print(" | ".join(str(v) for v in combo) + " | " + str(result))

# Function to convert logic expression to natural language
def build_sentence(expr):
    expr_sentence = expr
    for var, text in sentences.items():
        expr_sentence = expr_sentence.replace(var, text)
    expr_sentence = expr_sentence.replace("^", " and ")
    expr_sentence = expr_sentence.replace("v", " or ")
    expr_sentence = expr_sentence.replace("~", "not ")
    return expr_sentence

# Main menu loop
while True:
    print("\n--- First Order Logic Menu ---")
    print("1. Add a sentence (variable mapping)")
    print("2. Generate truth table from expression")
    print("3. Convert logic expression to sentence")
    print("4. Exit")
    
    choice = input("Select an option (1-4): ")
    
    if choice == "1":
        var = input("Enter variable name (e.g., P): ").strip()
        text = input(f"Enter sentence for {var}: ").strip()
        sentences[var] = text
        print(f"Mapping added: {var} = '{text}'")
        
    elif choice == "2":
        if not sentences:
            print("Please add at least one sentence first!")
            continue
        expr = input("Enter logical expression (use ^ for AND, v for OR, ~ for NOT): ").strip()
        # Extract only the variables (letters) and ignore operators
        variables = sorted(set([c for c in expr if c.isalpha()]))
        truth_table(expr, variables)
        
    elif choice == "3":
        if not sentences:
            print("Please add at least one sentence first!")
            continue
        expr = input("Enter logical expression: ").strip()
        print("Sentence form:", build_sentence(expr))
        
    elif choice == "4":
        print("Exiting program...")
        break
        
    else:
        print("Invalid option! Please select again.")
```
