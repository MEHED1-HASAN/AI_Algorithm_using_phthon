```python
from collections import deque

def depth_limited_search(graph, start, goal, limit):
    stack = deque()
    stack.append((start, 0))  # store (node, current_depth)
    visited = set()

    print(f"Depth-Limited Search (limit = {limit}):")

    while stack:
        current_node, depth = stack.pop()

        if current_node not in visited:
            print(f"{current_node} (depth {depth})")
            visited.add(current_node)

            if current_node == goal:
                print(f"Goal '{goal}' found at depth {depth}.")
                return

            if depth < limit:
                for neighbor in reversed(graph.get(current_node, [])):
                    if neighbor not in visited:
                        stack.append((neighbor, depth + 1))

    print(f"Goal '{goal}' not found within depth limit {limit}.")

print("Enter the graph in Python dictionary format (e.g., {'A': ['B', 'C'], 'B': ['D']}):")
graph = eval(input())

start_node = input("Enter the start node: ")
goal_node = input("Enter the goal node: ")
depth_limit = int(input("Enter the depth limit: "))

depth_limited_search(graph, start_node, goal_node, depth_limit)
```

```python
def depth_limited_search(graph, start, limit, goal):
    visited = set()
    stack = []
    traversal_order = []# list as stack

    # Each item in stack is a tuple: (node, current_depth)
    stack.append((start, 0))
    visited.add(start)
    
    goal_found = False
    print(f"Depth-Limited Search (limit={limit}):")

    while stack:
        current_node, depth = stack.pop()   # LIFO
        traversal_order.append(current_node)

        # Check if we reached the goal
        if current_node == goal:
            print(f"\nGoal '{goal}' found at depth {depth}.")
            
            goal_found = True

        # Only explore neighbors if we haven't reached the depth limit
        if depth < limit:
            # add neighbours in reverse order for consistent left-to-right traversal
            for neighbor in reversed(graph[current_node]):
                if neighbor not in visited:
                    visited.add(neighbor)
                    stack.append((neighbor, depth + 1))  # increment depth
    print(" ".join(traversal_order))
    if  not goal_found:
        print(f"\nGoal '{goal}' not found within depth limit {limit}.")

# Example usage
print("Enter the graph in Python dictionary format (e.g., {'A':['B','C'], 'B':['D']}):")
graph = eval(input())

start=(input("Enter the start node:"))
goal = input("Enter the goal node: ")
limit = int(input("Enter depth limit: "))

depth_limited_search(graph, start, limit, goal)
```
