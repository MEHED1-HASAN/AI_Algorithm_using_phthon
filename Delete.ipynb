```python
from collections import deque

def bidirectional_bfs(graph, start, goal):
    if start == goal:
        print("Start and goal are the same:", start)
        return

    visited_start = set([start])
    visited_goal = set([goal])

    queue_start = deque([start])
    queue_goal = deque([goal])

    while queue_start and queue_goal:
        # Forward BFS step
        curr_start = queue_start.popleft()
        for neighbor in graph.get(curr_start, []):
            if neighbor in visited_goal:
                print("Path found via:", neighbor)
                return
            if neighbor not in visited_start:
                visited_start.add(neighbor)
                queue_start.append(neighbor)

        # Backward BFS step
        curr_goal = queue_goal.popleft()
        for neighbor in graph.get(curr_goal, []):
            if neighbor in visited_start:
                print("Path found via:", neighbor)
                return
            if neighbor not in visited_goal:
                visited_goal.add(neighbor)
                queue_goal.append(neighbor)

    print("No path found.")

 # Example: A:B,C B:D,E C:F D: E: F:
graph =eval(input())

start_node = input("Start node: ")
goal_node = input("Goal node: ")

bidirectional_bfs(graph, start_node, goal_node)
```

```python
def UCS(graph,start,goal):
    visited=set()
    queue=[(0,start,[start])]

    while queue:
        queue.sort()
        cost,current,path=queue.pop(0)

        if current==goal:
            print(f"Goal found at cost {cost}")
            print("Path:","-->".join(path))
            return

        if current not in visited:
            visited.add(current)
        for neighbour,u_cost in graph[current]:
            if neighbour not in visited:
                new_cost=cost+u_cost
                new_path=path+[neighbour]
                queue.append((new_cost,neighbour,new_path))

print("Enter the graph with costs:")
graph = eval(input())
start = input("Enter start node: ")
goal = input("Enter goal node: ")

UCS(graph, start, goal)
```
