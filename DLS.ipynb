```python
from collections import deque

def depth_limited_search(graph, start, goal, limit):
    stack = deque()
    stack.append((start, 0))  # store (node, current_depth)
    visited = set()

    print(f"Depth-Limited Search (limit = {limit}):")

    while stack:
        current_node, depth = stack.pop()

        if current_node not in visited:
            print(f"{current_node} (depth {depth})")
            visited.add(current_node)

            if current_node == goal:
                print(f"Goal '{goal}' found at depth {depth}.")
                return

            if depth < limit:
                for neighbor in reversed(graph.get(current_node, [])):
                    if neighbor not in visited:
                        stack.append((neighbor, depth + 1))

    print(f"Goal '{goal}' not found within depth limit {limit}.")

print("Enter the graph in Python dictionary format (e.g., {'A': ['B', 'C'], 'B': ['D']}):")
graph = eval(input())

start_node = input("Enter the start node: ")
goal_node = input("Enter the goal node: ")
depth_limit = int(input("Enter the depth limit: "))

depth_limited_search(graph, start_node, goal_node, depth_limit)
```

```python
def DLS(graph, start, goal, limit):
    visited = set()
    stack = [(start, 0)]
    visited.add(start)
    goal_found = False
    goal_depth = None
    print(f"DLS with depth limit {limit}:")
    
    while stack:
        current_node, depth = stack.pop()
        print(current_node, end=" ")
        
        if current_node == goal and not goal_found:
            goal_found = True
            goal_depth = depth
            # Don't return yet - continue exploring
            
        if depth < limit:  # Expand only within depth limit
            for neighbour in reversed(graph[current_node]):
                if neighbour not in visited:
                    visited.add(neighbour)
                    stack.append((neighbour, depth + 1))
    
    # After exploring all nodes within depth limit
    if goal_found:
        print(f"\nGoal found at depth {goal_depth}")
        return True
    else:
        print(f"\nGoal not found within depth limit {limit}")
        return False

# Example usage:
print("Enter the graph (e.g., {'A':['B','C'], 'B':['D','E'], 'C':['F'], 'D':[], 'E':[], 'F':[]}):")
graph = eval(input())
start = input("Enter the start node: ")
goal = input("Enter the goal node: ")
limit = int(input("Enter the depth limit: "))

DLS(graph, start, goal, limit)
```
