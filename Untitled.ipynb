```python
def logic():
    s1=input("Enter the first statement:")
    s2=input("Enter the second statement:")
    p=input(f"Is '{s1}' TRUE or FALSE?")=='TRUE'
    q=input(f"Is '{s2}' TRUE or FALSE?")=='TRUE'
    result=None
    sentence=""
    ope=input("Enter any operation(Negation,conjunction,disjunction,implication,biconditional):").lower()
    if ope=='negation':
        result=not p
        sentence=f"~{s1}"
    elif ope=='conjunction':
        result=p and q
        sentence=f"{s1}^{s2}"
    elif ope=='disjunction':
        result=p or q
        sentence=f"{s1}V{s2}"
    elif ope=='implication':
        result=(not p) or q
        sentence=f"{s1}-->{s2}"
    elif ope=='biconditional':
        result= p==q
        sentence=f"{s1}<--->{s2}"
    else:
        print("Invalid operation input!")

    print(result)
    print(sentence)


logic()
        
    
    
```

```python
def super_simple_logic():
    print("Super Simple Logic")
    
    # Ask for quantifier or basic logic
    choice = input("Quantifier (Q) or Basic Logic (B)? ").upper()
    
    if choice == "Q":
        # Quantifier logic
        quant = input("SOME or EVERY? ").upper()
        prop = input("Property (e.g., 'is tall'): ")
        n = int(input("How many objects? "))
        
        result = True if quant == "EVERY" else False
        
        for i in range(1, n + 1):
            ans = input(f"Object {i} has '{prop}'? (y/n): ").lower() == "y"
            
            if quant == "EVERY":
                result = result and ans
            else:  # SOME
                result = result or ans
        
        print(f"{quant} object has '{prop}': {result}")
        
    else:
        # Basic logic
        p = input("First statement true? (y/n): ").lower() == "y"
        q = input("Second statement true? (y/n): ").lower() == "y"
        
        print(f"NOT first: {not p}")
        print(f"First AND second: {p and q}")
        print(f"First OR second: {p or q}")

# Run it
super_simple_logic()
```

```python
from collections import deque
def BFS(graph,start):
    visited=set()
    queue=deque()
    visited.add(start)
    queue.append(start)
    while queue:
        current=queue.popleft()
        print(current , end=" ")

        for neighbour in graph[current]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)

graph=eval(input("Enter the graph:"))
start=input("Enter the start Node:")
BFS(graph,start)
```

```python
def DFS(graph,start):
    visited=set()
    stack=[]
    visited.add(start)
    stack.append(start)

    while stack:
        current=stack.pop()
        print(current,end=" ")

        for neighbour in reversed(graph[current]):
            if neighbour not in visited:
                visited.add(neighbour)
                stack.append(neighbour)
graph=eval(input("Enter the graph:"))
start=input("Enter the starting node:")
DFS(graph,start)
```

```python
def DLS(graph,start,goal,limit):
    visited=set()
    stack=[(start,0)]
    visited.add(start)
    goal_found=False
    goal_dept=None
    while stack:
        current,depth=stack.pop()
        print(current, end=" ")

        if current==goal and not goal_found:
            goal_found=True
            goal_depth=depth
        if depth<limit:
            for neighbour in reversed(graph[current]):
                if neighbour not in visited:
                    visited.add(neighbour)
                    stack.append((neighbour, depth+1))
    if goal_found:
        print(f"Goal is found at depth {goal_dept}")
    else:
        print("Goal Not found")

graph=eval(input("Enter the graph:"))
start=input("Enter the start node:")
goal=input("enter the goal node:")
limit=int(input("Enter the limit:"))
DLS(graph,start,goal,limit)
```

```python
def IDDFS(graph,start,goal,maxi):
    for i in range(maxi+1):
        print(f"Level {i}:",end=" ")
        visited=set()
        stack=[]
        stack.append((start,0))
        visited.add(start)

        while stack:
            current,depth=stack.pop()
            print(current,end=" ")

            if current==goal:
                print("Goal found at level",depth)
                return True

            if depth<i:
                for neighbour in reversed(graph[current]):
                    if neighbour not in visited:
                        visited.add(neighbour)
                        stack.append((neighbour,depth+1))
        print("Goal Not found at level",i)


    print("Maximum level reached")

graph=eval(input("Enter the graph:"))
start=input("Enter the start node:")
goal=input("Enter the goal node:")
maxi=int(input("enter the maximum limit:"))
IDDFS(graph,start,goal,maxi)   
```

```python
def UCS(graph,start,goal):
    visited=set()
    stack=[(0,start,[start])]
    while stack:
         stack.sort()
         cost,current,path=stack.pop(0)
         print("Path:","-->".join(path))

         if current==goal:
             print("Goal found at cost:",cost)
             return

         if current not in visited:
             visited.add(current)

         for neighbour,up_cost in graph[current]:
             if neighbour not in visited:
                 new_cost=cost+up_cost
                 new_path=path+[neighbour]
                 stack.append((new_cost,neighbour,new_path))

print("Enter the graph with costs:")
graph = eval(input())
start = input("Enter start node: ")
goal = input("Enter goal node: ")

UCS(graph, start, goal)




    
```
