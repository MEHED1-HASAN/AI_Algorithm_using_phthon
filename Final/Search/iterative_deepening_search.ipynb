```python
from collections import deque
import ast

def iterative_deepening_dfs(graph, start, target, max_depth):
    print(f"IDDFS Traversal to find '{target}':")
    
    for depth_limit in range(max_depth + 1):
        print(f"\nDepth limit = {depth_limit}:")
        visited = set()
        stack = deque()
        stack.append((start, 0))  # (node, depth)

        found = False
        while stack:
            current_node, depth = stack.pop()
            if current_node not in visited:
                print(current_node, end=" ")
                visited.add(current_node)

                if current_node == target:
                    print(f"\nTarget '{target}' found at depth {depth_limit}.")
                    return

                if depth < depth_limit:
                    for neighbor in reversed(graph.get(current_node, [])):
                        if neighbor not in visited:
                            stack.append((neighbor, depth + 1))
    print(f"\nTarget '{target}' not found within depth {max_depth}.")

# Input Section
print("Enter the graph in Python dictionary format (e.g., {'A': ['B', 'C'], 'B': ['D']}):")
graph = ast.literal_eval(input())

start_node = input("Enter the start node: ")
target_node = input("Enter the target node to search: ")
max_depth = int(input("Enter the maximum depth limit: "))

iterative_deepening_dfs(graph, start_node, target_node, max_depth)
```
