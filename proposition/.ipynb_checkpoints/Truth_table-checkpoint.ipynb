```python
def logic_truth_table():
    stmt1 = input("Enter first statement (e.g., 'Raju is Doctor'): ")
    stmt2 = input("Enter second statement (e.g., 'Raju is Honest'): ")

    values = [True, False]

    print("\nTruth Table for:", "P=",stmt1, "\nQ=", stmt2)
    print("-"*90)
    print(f"P     Q     ~P        P ^ Q    P V Q    P -> Q   P <-> Q")
    print("-"*90)

    for P in values:
        for Q in values:
            neg_p = not P
            conj = P and Q
            disj = P or Q
            impl = (not P) or Q
            bicond = (P == Q)

            print(f"{str(P):8} {str(Q):8} {str(neg_p):8} {str(conj):8} {str(disj):8} {str(impl):8} {str(bicond):8}")

logic_truth_table()
```
